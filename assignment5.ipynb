```python
#base class
class House:
    def __init__(self, owner, location, rooms, price):
        self.owner = owner
        self.location = location
        self.rooms = rooms
        self.__price = price #this attribute has been encapsulated as it is a private attribute

#method to show house details
    def show_info(self):
        return f"{self.owner}'s house in {self.location} with {self.rooms} rooms "

#pricing of the House considering price is a private attribute
    def get_price(self):
         return f"The price of the house is ${self.__price}."

#method after house renovation new house pricing
    def renovate(self, added_value):
         self.__price += added_value
         return f"The house value increased to #{self.__price} after renovation."

#derived class inheritance
class Mansion (House):
    def __init__(self, owner, location, rooms, price, luxury_level):
        super().__init__(owner, location, rooms, price)
        self.luxury_level = luxury_level
 
    def show_info(self):
        return f"{self.owner}'s luxurious mansion in {self.location} with {self.rooms} rooms and {self.luxury_level} luxury level."

#creating objects
House1 = House ("Faith", "Lavington", 4, 80000, )
Mansion1 = Mansion("Mukami", "Karen", 10, 500000, "High")

print(House1.show_info())
print(House1.get_price())
print(House1.renovate(10000))
print()
print(Mansion1.show_info())
print(Mansion1.get_price())
```

```python
# Base class
class Vehicle:
    def move(self):
        print("The vehicle is moving...")

class Car(Vehicle):
    def move(self):
        print(" The car is driving on the road.")

class Plane(Vehicle):
    def move(self):
        print(" The plane is flying in the sky.")

class Boat(Vehicle):
    def move(self):
        print(" The boat is sailing on the water.")

class Bicycle(Vehicle):
    def move(self):
        print(" The bicycle is pedaling along the path.")

# Create objects
vehicles = [Car(), Plane(), Boat(), Bicycle()]
# Demonstrate polymorphism
for v in vehicles:
    v.move()
```
